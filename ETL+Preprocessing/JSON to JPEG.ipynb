```python
import cv2
import numpy as np
from base64 import b64decode
import json
from matplotlib import pyplot as plt

with open('/content/rekog-store') as f:
    d = json.load(f)

JPEG_r = b64decode(d["image"])

na_r = cv2.imdecode(np.frombuffer(JPEG_r,dtype=np.uint8), cv2.IMREAD_COLOR)
na_r = cv2.cvtColor(np.array(na_r), cv2.COLOR_BGR2RGB)

plt.imshow(na_r)
plt.show()
```
